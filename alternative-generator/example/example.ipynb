<h3>Примеры распределений</h3>
<p>'normal' - нормальное распределение</p>
<p>'peak' - узкий пик в центре графика плотности распределения через beta распределение</p>
<p>'uneven' - неравномерное распределение через сумму beta распределений с случайными параметрами</p>
<p>'uniform' - равномерное распределение</p>
<h3>Неравномерное распределений</h3>
<p>'uneven' распределение - формируется неравномерное распределение через итеративную генерацию beta распределения с случайными параметрами a и b. Можно задавать границы генерации a и b, а также количество итераций. Значения по-умолчанию получены экспериментально как наиболее оптимальные для неравномерного распределения.</p>
<p>Параметры:</p>
<p>- a_min - минимальная граница a для итераций beta-распределения</p>
<p>- a_max - максимальная граница a для итераций beta-распределения</p>
<p>- b_min - минимальная граница b для итераций beta-распределения</p>
<p>- b_max - максимальная граница b для итераций beta-распределения</p>
<p>- iterations - количество итераций. Должно быть делителем для количества альтернатив, иначе размеры данных могут не совпасть</p>

```python
def generate_and_show_data(min_value = 0, max_value = 1000, num_alternatives = 1000, a_min = 0, a_max = 60, b_min = 0, b_max = 60, iterations = 10, distribution='uneven'):
    if distribution == 'normal':
        data = np.random.normal(loc=(max_value + min_value) / 2,
                                scale=(max_value - min_value) / 6,
                                size=num_alternatives)
    elif distribution == 'peak':
        a, b = 50, 50  # Чем больше, тем уже пик
        data = np.random.beta(a, b, size=num_alternatives)
        data = data * (max_value - min_value) + min_value  # Масштабирование
    elif distribution == 'uneven':
        all_values = []
        for i in range(iterations):
            peak = np.random.beta(a=np.random.uniform(a_min, a_max), b=np.random.uniform(b_min, b_max), size=int(num_alternatives / iterations)) * (max_value - min_value) + min_value
            all_values.append(peak)
        data = np.concatenate(all_values)
        np.random.shuffle(data)
    elif distribution == 'uniform':
        data = np.random.uniform(min_value, max_value, size=num_alternatives)
    else:
        raise ValueError

    # Ограничение значений
    data = np.clip(data, min_value, max_value)
    print(type(data))
    # Построение графиков

    plt.figure(figsize=(7, 4))
    plt.plot(data, marker='o', linestyle='-', label='Random Data')
    plt.title(f'График сгенерированных данных ({distribution})')
    plt.xlabel('Индекс')
    plt.ylabel('Значение')
    plt.grid(True)
    plt.legend()
    plt.show()

    plt.figure(figsize=(7, 4))
    plt.hist(data, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(f'Распределение данных ({distribution})')
    plt.xlabel('Значение')
    plt.ylabel('Частота')
    plt.grid(True)
    plt.show()
```

```python
generate_and_show_data(distribution='uneven')
```

```python
generate_and_show_data(distribution='normal')
```

```python
generate_and_show_data(distribution='uniform')
```

```python
generate_and_show_data(distribution='peak')
```
